## Histograma de una imagen

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['image.cmap'] = 'gray'
%matplotlib inline
from skimage import io
from skimage import color
```

```python
# Funcion para leer una imagen y guardarla en una varible
I = io.imread("Flowers.jpg")
# Funcion para mostrar la imagen
plt.imshow(I)
```

```python
# Funcion para obtener el histograma de una imagen
def ImHist(Image):
    """
    Calculate the histogram of a grayscale image
    
    Parameters
    ----------------
    Image : The function receives a grayscale image of N * M dimensions
    
    Returns
    ----------------
    ImageHistogram : The function returns the histogram of the image on a scale of 0 to 255
    """
    
    I = np.uint8(Image)
    In = np.asarray(I, dtype = np.uint8)
    [r, c] = I.shape
    H = np.zeros((256, 1))
    
    for i in range (r):
        for j in range (c):
            k = I[i, j]
            H[k] = H[k] + 1
            
    ImageHistogram = H
    plt.stem(ImageHistogram, linefmt = 'k-', markerfmt = 'none', basefmt = 'r-')
    plt.title("Histogram")
    

# Funcion para convertir una imagen a escala de grises
def GscAverage(Image):
    """
    Conversion of a RGB image to a gray image by the Average method
    
    Parameters
    ----------------
    Image : The function receives a grayscale image of N * M dimensions
    
    Returns
    ----------------
    IOUT : The function returns a grayscale image by the average method
    """
    
    Im = Image
    [r, c, ch] = Im.shape
    IOUT = np.zeros((r, c), dtype = np.uint8)
    
    ImR = np.double(Im[:,:,0])
    ImG = np.double(Im[:,:,1])
    ImB = np.double(Im[:,:,2])
    
    IOUT = np.uint8((ImR + ImG + ImB)/ch)
    
    return IOUT
```

```python
II = GscAverage(I)
ImHist(II)
```
